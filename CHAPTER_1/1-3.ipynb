```python
# 웹 서버의 액세스 로그의 예

import re
import pandas as pd

#로그의 각 행에 매치하는 정규 표현
pattern = re.compile('^\S+ \S+ \S+ \[(.*)\] "(.*)" (\S+) (\S+)$')

def parse_access_log(path):
    for line in open(path, 'rt', encoding='UTF8'):
        for m in pattern.finditer(line):
            yield m.groups()

columns = ['time', 'request', 'status', 'bytes']
pd.DataFrame(parse_access_log("bigdata_jpub\\CHAPTER_1\\access.log"), columns = columns)
```

```python
df = pd.DataFrame(parse_access_log('bigdata_jpub\\CHAPTER_1\\access.log'), columns = columns)
df.time = pd.to_datetime(df.time, format='%d/%b/%Y:%X', exact=False)
df.head(2)

```

```python
df.to_csv('bigdata_jpub\\CHAPTER_1\\access_log.csv', index=False)
open("bigdata_jpub\\CHAPTER_1\\access_log.csv", 'rt', encoding='UTF-8').readlines()[:4]

```
